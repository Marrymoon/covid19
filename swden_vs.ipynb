```python
import pandas as pd
```

```python
import datetime
from bokeh.plotting import figure
from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.layouts import gridplot
```

```python
path = 'data/sweden_vs.csv'
DF = pd.read_csv(path)
DF['date'] = pd.to_datetime(DF['date'])
```

```python
DF.head()
df_seattle = DF[DF['region'] == 'Seattle']
df_belgium = DF[DF['region'] == 'Belgium']
df_sweden = DF[(DF.region == 'Sweden') & (DF.deaths > 0)]
df_norway = DF[(DF.region == 'Norway') & (DF.deaths > 0)]
df_ny = DF[(DF.region == 'New York') & (DF.deaths > 0)]
```

```python
def incidents_over_time_bar(df, key, window= 3, plot_height = 600, 
             plot_width = 600, title = None, line_width = 5, y_range = None): 
    labels = df['date'].tolist()
    if isinstance(labels[0], datetime.date):
        labels = [datetime.datetime(x.year, x.month, x.day) for x in labels]
    nums = df[key].rolling(window).mean()
    p = figure(x_axis_type = 'datetime', title = title, 
                 plot_width = plot_width , plot_height = plot_height, y_range = y_range)
    p.vbar(x=labels, top=nums, line_width = line_width, width = .9) 
    return p
```

```python
output_notebook()
def do_graphs(df_sweden, df_other, pop_other, title, l, pop_sweden = 10.2):
    df_sweden['deaths_adj'] = df_sweden['deaths']* (pop_other/pop_sweden)
    max1 =max(df_sweden['deaths_adj'].tolist())
    max2 =max(df_other['deaths'].tolist())
    the_max = max(max1, max2)
    p_other = incidents_over_time_bar(df = df_other, key = 'deaths', title = title,
                                 y_range = (0, the_max))
    p_sweden = incidents_over_time_bar(df = df_sweden, key = 'deaths_adj', title = 'Sweden',
                                  y_range = (0, the_max))
    l.append(p_sweden)
    l.append(p_other)
    return l

#pop Sweden 10.2
# norway 5.4
#pop King 2.25
# pop NY 8.4
# pop Belgium 11.5
l = []
l = do_graphs(df_sweden, df_norway, pop_other = 5.4, pop_sweden = 10.2, title = 'Norway', l =l)
l = do_graphs(df_sweden, df_seattle, pop_other = 2.3, pop_sweden = 10.2, title = 'Seattle', l =l)
l = do_graphs(df_sweden, df_ny, pop_other = 8.4, pop_sweden = 10.2, title = 'New York', l =l)
l = do_graphs(df_sweden, df_ny, pop_other = 11.5, pop_sweden = 10.2, title = 'Belgium', l =l)

grid = gridplot(l, ncols = 2)
show(grid)
len(l)
```
