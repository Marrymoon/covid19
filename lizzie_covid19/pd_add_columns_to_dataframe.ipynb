```python
import pandas as pd

# List of Tuples
students = [('jack', 34, 'Sydeny', 'Australia'),
            ('Riti', 30, 'Delhi', 'India'),
            ('Vikas', 31, 'Mumbai', 'India'),
            ('Neelu', 32, 'Bangalore', 'India'),
            ('John', 16, 'New York', 'US'),
            ('Mike', 17, 'las vegas', 'US')]

# Add column to Dataframe in Pandas using operator []
# Create a DataFrame object
df_obj = pd.DataFrame(students,
                      columns=['Name', 'Age', 'City', 'Country'],
                      index=['a', 'b', 'c', 'd', 'e', 'f'])

df_obj
```

```python
# Pandas: Add new column to Dataframe with values in a list 
# Add column with Name Marks
    
df_obj['Marks'] =  [10, 20, 45, 33, 22,11]

df_obj
```

```python
# Add new column to DataFrame with same default value
df_obj['Total'] = 50

df_obj
```

```python
# Add new column in DataFrame with values based on other columns

df_obj['Percentage'] = (df_obj['Marks'] / df_obj['Total'] * 100)

df_obj
```

```python


# Create a DataFrame object 
df_obj1 = pd.DataFrame(students,
 columns=['Name', 'Age', 'City', 'Country'],
 index=['a', 'b', 'c', 'd', 'e', 'f'])

df_obj1
```

```python
# Append column to dataFrame using assign() function

mod_fd = df_obj1.assign(Marks=[10, 20, 45, 33, 22, 11])
mod_fd
```

```python
# Add multiple columns in DataFrame using assign()

df_obj2 = df_obj1.assign(Marks=[10, 20, 45, 33, 22, 11], Total=[50]*6)
df_obj2
```

```python
# Add a columns in DataFrame based on other column using lambda function

df_obj3 = df_obj2.assign(Percentage = lambda x: (x['Marks']/x['Total'] * 100))

df_obj3
```

```python
# Insert new column to Dataframe using insert()

df_obj3 = pd.DataFrame(students,
 columns=['Name', 'Age', 'City', 'Country'],
 index=['a', 'b', 'c', 'd', 'e', 'f'])

# Insert column at the 2nd position of Dataframe
df_obj3.insert(3, "Marks", [10, 20, 45, 33, 22, 11], True)

df_obj3
```

```python
#  Add a column to Dataframe by dictionary

ids = [21,43,50,56,60,98]

df_obj3['ID'] = dict(zip(ids, df_obj3['Name']))

df_obj3
```
