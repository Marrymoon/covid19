```python
import pandas as pd
import copy
```

```python
import datetime
from bokeh.plotting import figure
from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.layouts import gridplot
```

```python
path = 'data/states.csv'
DF = pd.read_csv(path)
DF['date'] = pd.to_datetime(DF['date'])
DF_POP = pd.read_csv('data/states_population.csv')
DF_POP.head()
```

```python
def incidents_over_time_bar(df, key, window= 3, plot_height = 600, 
             plot_width = 600, title = None, line_width = 5, y_range = None,
                           x_range=None): 
    labels = df['date'].tolist()
    if isinstance(labels[0], datetime.date):
        labels = [datetime.datetime(x.year, x.month, x.day) for x in labels]
    nums = df[key].rolling(window).mean()
    p = figure(x_axis_type = 'datetime', title = title, 
                 plot_width = plot_width , plot_height = plot_height, y_range = y_range,
              x_range = x_range)
    p.vbar(x=labels, top=nums, line_width = line_width, width = .9) 
    return p
```

```python
output_notebook()
```

```python
def shape_data(state):
    df = DF[DF['state'] == state]
    pop = DF_POP[DF_POP['state'] == state]['population_2019'].tolist()
    df['deaths_pop'] = df['deaths'] * (1e6 / pop[0] )
    df['cases_pop'] = df['cases'] * (1e6 / pop[0] )
    return df

df_wash = shape_data('Washington')
df_ny = shape_data('New York')
df_arizona = shape_data('Arizona')
df_texas = shape_data('Texas')
df_arkansas = shape_data('Arkansas')
df_florida = shape_data('Florida')


l = []
l.append(incidents_over_time_bar(df_arizona, 'deaths_pop', plot_width = 300, plot_height = 300,
                                   title = 'Arizona'))
l.append(incidents_over_time_bar(df_wash, 'deaths_pop', plot_width = 300, plot_height = 300,
                                title = 'Washington'))
l.append(incidents_over_time_bar(df_ny, 'deaths_pop', plot_width = 300, plot_height = 300, 
                              title = 'New York'))
l.append(incidents_over_time_bar(df_texas, 'deaths_pop', plot_width = 300, plot_height = 300,
                                 title = 'Texas'))
l.append(incidents_over_time_bar(df_wash, 'deaths_pop', plot_width = 300, plot_height = 300,
                                title = 'Washington'))
l.append(incidents_over_time_bar(df_ny, 'deaths_pop', plot_width = 300, plot_height = 300, 
                              title = 'New York'))
l.append(incidents_over_time_bar(df_arkansas, 'deaths_pop', plot_width = 300, plot_height = 300, 
                              title = 'Arkansas'))
l.append(incidents_over_time_bar(df_wash, 'deaths_pop', plot_width = 300, plot_height = 300,
                                title = 'Washington'))
l.append(incidents_over_time_bar(df_ny, 'deaths_pop', plot_width = 300, plot_height = 300, 
                              title = 'New York'))
grid = gridplot(l, ncols = 3)
show(grid)
```

```python
def append_to_graph(df, l, title, keyword = 'cases_pop'):
    l.append(incidents_over_time_bar(df, keyword, plot_width = 300, 
    plot_height = 300, title = title, y_range = (0,600),
    x_range = (datetime.datetime(2020,3, 1) , datetime.datetime(2020,6, 20))                                                    
    ))

l = []
info = [(df_arizona, 'Arizona'), 
       (df_texas, 'Texas'),
        (df_arkansas, 'Arkansas'),
        (df_florida, 'Forida'),
        (shape_data('California'), 'California')
       ]
for i in info:
    append_to_graph(i[0], l, i[1])
    append_to_graph(df_wash, l, 'Washington')
    append_to_graph(df_ny, l, 'New york')
grid = gridplot(l, ncols = 3)
show(grid)
```
