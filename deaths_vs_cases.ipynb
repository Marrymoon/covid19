```python
import pandas as pd
import numpy as np
```

```python
import datetime
from bokeh.plotting import figure
from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.layouts import gridplot
```

```python
import statsmodels.formula.api as smf
```

```python
path = 'data/states_cum_pop.csv'
DF = pd.read_csv(path)
DF['cases_mil'] = DF['cases']/DF['population_2019'] * 1e6
DF['deaths_mil'] = DF['deaths']/DF['population_2019'] * 1e6
df = DF[(DF['deaths_mil'] > 10) ]

#DF['date'] = pd.to_datetime(DF['date'])
DF.head()
DF[DF['state'] == 'Washington']

```

```python
output_notebook()
```

```python
y = df['deaths_mil'].tolist()
x  = df['cases_mil'].tolist()
model = smf.ols(formula='deaths_mil ~ cases_mil' , data=DF)
res = model.fit()
a, b1 = res.params
print(a, b1)
y_fit = [a + b1* x for x in x]
p_value = res.pvalues
print(p_value)
```

```python
p = figure(title = 'deaths vs cases')

p.circle(x = x , y = y)
p.line(x = x, y = y_fit)
show(p)
```
