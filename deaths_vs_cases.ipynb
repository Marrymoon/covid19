```python
import pandas as pd
import numpy as np
```

```python
import datetime
from bokeh.plotting import figure
from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.layouts import gridplot
```

```python
path = 'data/states_cum_pop.csv'
DF = pd.read_csv(path)
DF['cases_mil'] = DF['cases']/DF['population_2019'] * 1e6
DF['deaths_mil'] = DF['deaths']/DF['population_2019'] * 1e6
df = DF[(DF['deaths_mil'] > 10) ]

#DF['date'] = pd.to_datetime(DF['date'])
DF.head()

```

```python
output_notebook()
```

```python
p = figure(title = 'deaths vs cases')
x = df['deaths_mil'].tolist()
y  = df['cases_mil'].tolist()
p.circle(x = x , y = y)
show(p)
```
